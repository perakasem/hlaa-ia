```python
import numpy as np
import matplotlib.pyplot as plt

data = [[0.047272074, 0.009758432, 0.003964148], 
        [0.04873435, 0.002453182, 0.001983919], 
        [0.054947517, 0.003014982, 0.002310078]]

X = np.arange(3)
fig = plt.figure(figsize=(4, 3))
ax = fig.add_axes([0,0,1,1])
method = ['Backward Euler', 'P-C', 'RK2']
ax.bar(X + 0.00, data[0], color = 'orange', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'forestgreen', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'cornflowerblue', width = 0.25)

ax.set_ylabel('Runtime / s')
ax.set_xlabel('Numerical Method')
ax.set_title('Runtimes by Method')

plt.xticks(X + 0.25, method)
plt.grid(axis = 'x')

ax.legend(labels=['TC1', 'TC2', 'TC3'], title="Test Cases")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center', va = 'bottom')

data = [0.0503, 0.0051, 0.0028]

fig = plt.figure(figsize=(3, 3))
ax = fig.add_axes([0,0,1,1])
method = ['Backward Euler', 'P-C', 'RK2']

ax.set_ylabel('Runtime / s')
ax.set_xlabel('Numerical Method')
ax.set_title('Average Runtime by Method')
ax.bar(X + 0.00, data, color = 'cornflowerblue')

plt.xticks(X, method)
addlabels(X, data)
plt.grid(axis = 'x')
```
